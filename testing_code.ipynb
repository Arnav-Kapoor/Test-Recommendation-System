```python
train=pd.read_excel('Gen_AI Dataset.xlsx',sheet_name='Train-Set')
```

```python
train
```

```python
train.drop_duplicates(subset='Query')
```

```python
ground_truth = {}
for _, row in train.iterrows():
    query = row['Query']
    if query not in ground_truth:
        ground_truth[query] = []
    ground_truth[query].append(row['Assessment_url'])
```
