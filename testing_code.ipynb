```python
import requests
import json

API_URL = "http://localhost:8000/recommend" # change to your Render URL

def recommend(query: str) -> dict:
    response = requests.post(API_URL, json={"query": query})
    response.raise_for_status()
    return response.json()

def display_results(result: dict):
    print(f"Query: {result['query']}\n")
    print(f"Reasoning: {result['reasoning']}\n")
    print(f"{'─' * 60}")
    
    for i, rec in enumerate(result['recommendations'], start=1):
        print(f"\n[{i}] {rec['title']}")
        print(f"  URL:               {rec['url']}")
        print(f"  Description:       {rec['description']}")
        print(f"  Job Levels:        {', '.join(rec['job_levels'])}")
        print(f"  Languages:         {', '.join(rec['languages'])}")
        print(f"  Assessment Length: {rec['assessment_length']}")
        print(f"  Test Types:        {', '.join(rec['test_types'])}")
        print(f"  Remote Testing:    {rec['remote_testing']}")
        print(f"  Adaptive/IRT:      {rec['adaptive_irt']}")
        # print(f"  Relevance Score:   {rec['relevance_score']}")
        print(f"  Why it fits:       {rec['relevance_reason']}")

# ── Run a query ───────────────────────────────────────────────

# query = "I am hiring for Java developers who can also collaborate effectively with my business teams. Looking for an assessment that can be completed in 40 minutes."


# for query in queries[-]:
query=queries[-3]
print(query)
result = recommend(query)
results.append(result)
    # display_results(result)
```

```python
query_url_mapping={}
for result in results:
    query=result['query']
    # print(query)
    for recommend in result['recommendations']:
        if query not in query_url_mapping:
            query_url_mapping[query] = []
        query_url_mapping[query].append(recommend['url'])
```

```python
query_url_mapping
```

```python
output=[]
for query,urls in query_url_mapping.items():
    for url in urls:
        output.append(pd.DataFrame({"Query":[query],'Assessment_url':[url]}))
```

```python
output=pd.concat(output).drop_duplicates()
```

```python
output.to_csv('arnav_kapoor.csv',index=False)
```
